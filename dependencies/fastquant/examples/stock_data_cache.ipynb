```python
from fastquant import get_pse_data_cache, get_stock_data
```

```python
df = get_pse_data_cache()
df.tail()
```

```python
phisix_symbol = 'JFC'
yahoo_symbol = "GOOGL"
start_date = '2020-4-1'
end_date = '2020-4-14'
format = "oc"
```

```python
#query within cache
df = get_stock_data(phisix_symbol, 
                    start_date, 
                    end_date, 
                    source="phisix", 
                    format=format
                   )
df.tail()
```

```python
#query within cache
df = get_stock_data(phisix_symbol, 
                    start_date, 
                    end_date, 
                    source="phisix", 
                    #format=format
                   )
df.tail()
```

```python
#query within cache
df = get_stock_data(yahoo_symbol, 
                    start_date, 
                    end_date, 
                    source="yahoo", 
                    format=format
                   )
df.tail()
```
